```python
from src.hooked_model import HookedSparseGPT
import torch
torch.set_grad_enabled(False)
model = HookedSparseGPT.from_pretrained("jacobcd52/ss_d128_f1", dtype=torch.float32)
```

```python
from datasets import load_dataset
dataset = load_dataset("SimpleStories/SimpleStories", split="train")
```

```python
texts = dataset["story"][:16]
toks = model.to_tokens(texts)[:, :64].cuda()
```

```python
with torch.autocast(device_type="cuda", dtype=torch.bfloat16):
    loss = model(toks, return_type="loss")
loss
```
